```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
!pip3 install tabulate
# Specify paths to CSV files
datasets = {
    'zhvi': '/Users/owais/Desktop/project/zhvi.csv',
    'zori': '/Users/owais/Desktop/project/zori.csv',
    'for_sale': '/Users/owais/Desktop/project/for_sale.csv',
    'new_listings': '/Users/owais/Desktop/project/new_listings.csv',
    'new_pending': '/Users/owais/Desktop/project/new_pending.csv',
    'mlp': '/Users/owais/Desktop/project/mlp.csv',
    'price_cuts': '/Users/owais/Desktop/project/median-price-cuts.csv',
    'new_con_sales_count': '/Users/owais/Desktop/project/new_con_sales_count.csv',
    'new_con_sales_price': '/Users/owais/Desktop/project/new_con_sales_price.csv'
}

# Load datasets
loaded_data = {name: pd.read_csv(path) for name, path in datasets.items()}

# Reshape and filter datasets from 2018 onwards
def reshape_and_filter(df, date_start='2018-01-01', date_format='%Y-%m-%d'):
    df_long = pd.melt(df, id_vars='RegionName', var_name='Date', value_name='Value')
    df_long['Date'] = pd.to_datetime(df_long['Date'], format=date_format, errors='coerce')
    return df_long[df_long['Date'] >= pd.to_datetime(date_start)]

# Combine datasets on RegionName and Date
def combine_datasets(loaded_data):
    combined_df_list = []
    for name, df in loaded_data.items():
        print(f"Processing dataset: {name}")
        reshaped_df = reshape_and_filter(df, date_format='%Y-%m-%d')
        reshaped_df.rename(columns={'Value': f'{name}_value'}, inplace=True)
        combined_df_list.append(reshaped_df)
    combined_data = combined_df_list[0]
    for df in combined_df_list[1:]:
        combined_data = pd.merge(combined_data, df, on=['RegionName', 'Date'], how='outer')
    return combined_data

# Load and combine datasets
combined_data = combine_datasets(loaded_data)

print(combined_data.head())

```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
from tabulate import tabulate  # For organized table display

# Model evaluation and plot results
def plot_results(y_test, y_pred, model_name):
    plt.scatter(y_test, y_pred)
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.title(f"Actual vs. Predicted ({model_name})")
    plt.show()

# Function to evaluate and print metrics
def evaluate_model(y_test, y_pred, model_name):
    mae = mean_absolute_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    print(f'{model_name} - MAE: {mae}, RMSE: {rmse}')
    plot_results(y_test, y_pred, model_name)

import matplotlib.pyplot as plt

# Function to plot real vs predicted values
def plot_real_vs_predicted(y_test, y_pred, model_name):
    plt.figure(figsize=(10, 6))

    # Plot actual values
    plt.plot(range(len(y_test)), y_test, label='Actual Values', color='blue')

    # Plot predicted values
    plt.plot(range(len(y_pred)), y_pred, label=f'{model_name} Predicted Values', color='red')

    # Adding labels and title
    plt.title(f"Real vs Predicted Values - {model_name}")
    plt.xlabel("Data Points")
    plt.ylabel("Values")
    
    # Display legend
    plt.legend()
    
    # Show the plot
    plt.tight_layout()
    plt.show()

# Function to train and predict model, modified to plot predictions
def train_and_predict_model(model_choice, X_train, X_test, y_train, y_test, model_results):
    if model_choice == '1':
        model = RandomForestRegressor(n_estimators=100, random_state=42)
        model_name = 'Random Forest'
    elif model_choice == '2':
        model = Ridge(alpha=1.0)
        model_name = 'Ridge Regression'
    elif model_choice == '3':
        poly = PolynomialFeatures(degree=2)
        X_train_poly = poly.fit_transform(X_train)
        X_test_poly = poly.transform(X_test)
        model = LinearRegression()
        model_name = 'Polynomial Regression'
        model.fit(X_train_poly, y_train)
        y_pred = model.predict(X_test_poly)
        plot_real_vs_predicted(y_test, y_pred, model_name)
        return model_results
    else:
        print("Invalid selection. Please select a valid model.")
        return model_results
    
    # Train and predict
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Store the results
    model_results[model_name] = y_pred

    # Plot actual vs predicted values
    plot_real_vs_predicted(y_test, y_pred, model_name)

    return model_results



# Evaluate model performance and display results
def evaluate_and_display_results(y_test, y_pred, index, model_name):
    mae = mean_absolute_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    print(f'{model_name} - MAE: {mae}, RMSE: {rmse}')
    
    # Plot true vs predicted values
    plt.figure(figsize=(10, 6))
    plt.plot(y_test.index, y_test, label='True Values', color='blue')
    plt.plot(y_test.index, y_pred, label='Predicted Values', color='red')
    plt.xlabel("Date")
    plt.ylabel("Values")
    plt.title(f"True vs. Predicted Values ({model_name})")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    # Display table of Date, True Values, and Predicted Values
    result_df = pd.DataFrame({'Date': y_test.index, 'True Value': y_test, 'Predicted Value': y_pred})
    result_df['Date'] = result_df.index.strftime('%Y-%m')  # Format date for display
    
    print("\nTrue vs. Predicted Table:")
    print(tabulate(result_df, headers='keys', tablefmt='grid'))

# Evaluate model performance, store results, and return the model comparison results
def evaluate_and_store_results(y_test, y_pred, model_name, model_results):
    mae = mean_absolute_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)

    # Store the results
    model_results[model_name] = {'RMSE': rmse, 'MAE': mae, 'y_test': y_test, 'y_pred': y_pred}

    # Display the results
    print(f"{model_name} - MAE: {mae}, RMSE: {rmse}")
    
    # Return the model results
    return model_results
    
# Function to compare all models' predictions (if needed)
def plot_model_comparisons(model_results):
    plt.figure(figsize=(10, 6))

    for model_name, y_pred in model_results.items():
        plt.plot(y_pred, label=f'{model_name} Predictions')

    plt.title("Comparison of Model Predictions")
    plt.xlabel("Data Points")
    plt.ylabel("Predicted Values")
    plt.legend()
    plt.tight_layout()
    plt.show()

```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
import pandas as pd

# Function to fit and forecast using SARIMA
def sarima_model(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12)):
    model = SARIMAX(train_data, order=order, seasonal_order=seasonal_order)
    sarima_fit = model.fit(disp=False)
    return sarima_fit

# Forecast the future values
def future_forecast(model, steps=12):
    forecast = model.get_forecast(steps=steps)
    return forecast.predicted_mean

# Clean and prepare the data for SARIMA, including scaling
def prepare_data(data, region, column):
    # Filter data for the specific region and column
    region_data = data[data['RegionName'] == region][column].dropna()

    # Convert to numeric, forcing errors to NaN and then dropping them
    region_data = pd.to_numeric(region_data, errors='coerce').dropna()

    # Reshape data for scaling
    region_data = region_data.values.reshape(-1, 1)

    # Apply MinMax scaling
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(region_data)

    return scaled_data, scaler

# Example of applying SARIMA to ZHVI data for a specific region
train_data_scaled, scaler = prepare_data(combined_data, 'New York, NY', 'zhvi_value')

if len(train_data_scaled) > 0:  # Ensure there's enough data
    sarima_model_fit = sarima_model(train_data_scaled)
    future_predictions_scaled = future_forecast(sarima_model_fit)
    
    # Inverse transform the predictions and the training data back to original scale
    future_predictions = scaler.inverse_transform(future_predictions_scaled.reshape(-1, 1))
    train_data = scaler.inverse_transform(train_data_scaled)
    
    print(future_predictions)
else:
    print("Not enough data for the specified region and column.")


# Function to plot historical data and SARIMA forecast with date labels
def plot_historical_and_forecast(train_data, future_predictions, region, start_date='2018-01'):
    plt.figure(figsize=(10, 6))
    
    # Create a date range for the historical and forecasted data
    historical_dates = pd.date_range(start=start_date, periods=len(train_data), freq='ME')
    forecast_dates = pd.date_range(start=historical_dates[-1] + pd.DateOffset(months=1), periods=len(future_predictions), freq='ME')

    # Plot historical ZHVI data
    plt.plot(historical_dates, train_data, label='Historical ZHVI Data', color='blue')

    # Plot SARIMA forecast
    plt.plot(forecast_dates, future_predictions, label='SARIMA Forecast', color='red')
    
    # Adding labels and title
    plt.title(f"Historical ZHVI Data and SARIMA Forecast for {region}")
    plt.xlabel("Time (Months/Years)")
    plt.ylabel("ZHVI Value")
    
    # Display legend
    plt.legend()

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Show the plot
    plt.tight_layout()
    plt.show()

# Example call with your data
plot_historical_and_forecast(train_data, future_predictions, 'New York, NY', start_date='2018-01')

#Clean and prepare the data for SARIMA, ensuring numeric format
def prepare_sarima_data(data, region, column):
    # Filter data for the specific region and column
    region_data = data[data['RegionName'] == region][column].dropna()

    # Convert to numeric, forcing errors to NaN and then dropping them
    region_data = pd.to_numeric(region_data, errors='coerce').dropna()

    # Ensure it is cast as a numpy array for SARIMAX
    return np.asarray(region_data)

# SARIMA forecasting based on user input for region and year
def run_sarima_forecasting(data, region, target_column):
    # Filter data for the specific region
    region_data = data[data['RegionName'] == region][target_column].dropna()
    
    if region_data.empty:
        print(f"No data available for the region '{region}'")
        return

    # Ask for forecast horizon (years into the future)
    forecast_years = int(input("Enter the number of years into the future for forecasting: "))
    steps = forecast_years * 12  # Assume monthly data

    # Fit SARIMA model
    sarima_model_fit = sarima_model(region_data)

    # Forecast future values
    future_predictions = future_forecast(sarima_model_fit, steps=steps)

    # Plot historical data and predictions
    plot_sarima_predictions(region_data, future_predictions, region)
    
# Function to plot SARIMA predictions
def plot_sarima_predictions(train_data, future_predictions, region):
    plt.figure(figsize=(10, 6))
    
    # Create a date range for the historical and forecasted data
    historical_dates = pd.date_range(start=train_data.index[0], periods=len(train_data), freq='M')
    forecast_dates = pd.date_range(start=historical_dates[-1] + pd.DateOffset(months=1), periods=len(future_predictions), freq='M')

    # Plot historical data
    plt.plot(historical_dates, train_data, label='Historical Data', color='blue')

    # Plot SARIMA forecast
    plt.plot(forecast_dates, future_predictions, label='SARIMA Forecast', color='red')
    
    plt.title(f"SARIMA Forecast vs Historical Data for {region}")
    plt.xlabel("Time (Months/Years)")
    plt.ylabel("Values")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Function to fit and forecast using SARIMA
def sarima_model(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12)):
    model = SARIMAX(train_data, order=order, seasonal_order=seasonal_order)
    sarima_fit = model.fit(disp=False)
    return sarima_fit

# Function to fit and forecast using ARIMA
def arima_model(train_data, order=(1, 1, 1)):
    model = ARIMA(train_data, order=order)
    arima_fit = model.fit()
    return arima_fit

# Forecast future values
def future_forecast(model_fit, steps=12):
    forecast = model_fit.get_forecast(steps=steps)
    return np.asarray(forecast.predicted_mean)

# Prepare data for time series models
def prepare_data(data, region, column):
    region_data = data[data['RegionName'] == region][column].dropna()
    if region_data.empty:
        print(f"No data found for region: {region}")
        return None, None

    region_data = pd.to_numeric(region_data, errors='coerce').dropna()
    if len(region_data) == 0:
        print(f"No valid data for region: {region}")
        return None, None

    region_data = region_data.values.reshape(-1, 1)
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(region_data)

    return scaled_data, scaler

# Inverse transform predictions to the original scale
def inverse_transform_predictions(scaler, future_predictions_scaled):
    return scaler.inverse_transform(future_predictions_scaled.reshape(-1, 1))

# Plot predictions
def plot_predictions(train_data, future_predictions_sarima, future_predictions_arima, region):
    plt.figure(figsize=(10, 6))
    historical_dates = pd.date_range(start='2018-01', periods=len(train_data), freq='M')
    forecast_dates = pd.date_range(start=historical_dates[-1] + pd.DateOffset(months=1), periods=len(future_predictions_sarima), freq='M')

    plt.plot(historical_dates, train_data, label='Historical Data', color='blue')
    plt.plot(forecast_dates, future_predictions_sarima, label='SARIMA Forecast', color='red')
    plt.plot(forecast_dates, future_predictions_arima, label='ARIMA Forecast', color='green')

    plt.title(f"SARIMA vs ARIMA Forecasts for {region}")
    plt.xlabel("Time (Months/Years)")
    plt.ylabel("Values")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Calculate and display evaluation metrics
def evaluate_predictions(true_data, sarima_predictions, arima_predictions):
    mse_sarima = mean_squared_error(true_data, sarima_predictions)
    mse_arima = mean_squared_error(true_data, arima_predictions)

    mae_sarima = mean_absolute_error(true_data, sarima_predictions)
    mae_arima = mean_absolute_error(true_data, arima_predictions)

    print(f"SARIMA - MSE: {mse_sarima}, MAE: {mae_sarima}")
    print(f"ARIMA - MSE: {mse_arima}, MAE: {mae_arima}")

# Run both SARIMA and ARIMA models for comparison
def run_forecasting_comparison(data, region, target_column):
    region_data_scaled, scaler = prepare_data(data, region, target_column)
    if region_data_scaled is None:
        return

    forecast_years = int(input("Enter the number of years into the future for forecasting: "))
    steps = forecast_years * 12

    sarima_fit = sarima_model(region_data_scaled)
    future_predictions_sarima_scaled = future_forecast(sarima_fit, steps=steps)
    future_predictions_sarima = inverse_transform_predictions(scaler, future_predictions_sarima_scaled)

    arima_fit = arima_model(region_data_scaled)
    future_predictions_arima_scaled = future_forecast(arima_fit, steps=steps)
    future_predictions_arima = inverse_transform_predictions(scaler, future_predictions_arima_scaled)

    train_data = scaler.inverse_transform(region_data_scaled)

    plot_predictions(train_data, future_predictions_sarima, future_predictions_arima, region)

    evaluate_predictions(train_data[-steps:], future_predictions_sarima, future_predictions_arima)

# Example usage
region = input("Enter the region (e.g., 'New York, NY'): ")
target = input("Choose a target: 'zhvi_value' for home prices or 'zori_value' for rent prices: ")

run_forecasting_comparison(combined_data, region, target)
```

```python
def get_user_input():
    # Ask what the user wants to predict
    while True:
        prediction_type = input("Select what you want to predict: (1) House Prices (ZHVI), (2) Rent Prices (ZORI): ")
        if prediction_type == '1':
            return 'zhvi_value'
        elif prediction_type == '2':
            return 'zori_value'
        else:
            print("Invalid selection. Please choose either 1 or 2.")

def get_region_input():
    # Ask for the region
    region = input("Enter the region you want to analyze (e.g., 'New York, NY'): ")
    return region

def validate_data(combined_data, features, target):
    missing_columns = [col for col in features + [target] if col not in combined_data.columns]
    if missing_columns:
        print(f"Missing columns: {missing_columns}")
        return False
    return True

def train_and_compare_models(X_train, X_test, y_train, y_test, model_results):
    # Allow user to train multiple models
    while True:
        model_choice = input("Select a model: (1) Random Forest, (2) Ridge Regression, (3) Polynomial Regression, (4) Quit: ")
        if model_choice == '4':
            break
        model_results = train_and_predict_model(model_choice, X_train, X_test, y_train, y_test, model_results)

    # Plot comparison of all models
    plot_model_comparisons(model_results)

import matplotlib.pyplot as plt

# Function to plot real vs predicted values
def plot_real_vs_predicted(y_test, y_pred, model_name):
    plt.figure(figsize=(10, 6))

    # Plot actual values
    plt.plot(range(len(y_test)), y_test, label='Actual Values', color='blue')

    # Plot predicted values
    plt.plot(range(len(y_pred)), y_pred, label=f'{model_name} Predicted Values', color='red')

    # Adding labels and title
    plt.title(f"Real vs Predicted Values - {model_name}")
    plt.xlabel("Data Points")
    plt.ylabel("Values")
    
    # Display legend
    plt.legend()
    
    # Show the plot
    plt.tight_layout()
    plt.show()

# Function to train and predict model, modified to plot predictions
def train_and_predict_model(model_choice, X_train, X_test, y_train, y_test, model_results):
    if model_choice == '1':
        model = RandomForestRegressor(n_estimators=100, random_state=42)
        model_name = 'Random Forest'
    elif model_choice == '2':
        model = Ridge(alpha=1.0)
        model_name = 'Ridge Regression'
    elif model_choice == '3':
        poly = PolynomialFeatures(degree=2)
        X_train_poly = poly.fit_transform(X_train)
        X_test_poly = poly.transform(X_test)
        model = LinearRegression()
        model_name = 'Polynomial Regression'
        model.fit(X_train_poly, y_train)
        y_pred = model.predict(X_test_poly)
        plot_real_vs_predicted(y_test, y_pred, model_name)
        return model_results
    else:
        print("Invalid selection. Please select a valid model.")
        return model_results
    
    # Train and predict
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Store the results
    model_results[model_name] = y_pred

    # Plot actual vs predicted values
    plot_real_vs_predicted(y_test, y_pred, model_name)

    return model_results

# Function to compare all models' predictions (if needed)
def plot_model_comparisons(model_results):
    plt.figure(figsize=(10, 6))

    for model_name, y_pred in model_results.items():
        plt.plot(y_pred, label=f'{model_name} Predictions')

    plt.title("Comparison of Model Predictions")
    plt.xlabel("Data Points")
    plt.ylabel("Predicted Values")
    plt.legend()
    plt.tight_layout()
    plt.show()

# Main function to manage the overall process
def main():
    # Load data
    combined_data = combine_datasets(loaded_data)

    # Get user inputs
    target = get_user_input()
    region = get_region_input()

    # Features used for predictions
    features = ['for_sale_value', 'price_cuts_value', 'new_listings_value']

    # Filter data by region and check if sufficient data
    region_data = combined_data[combined_data['RegionName'] == region].dropna(subset=features + [target])
    if region_data.empty:
        print(f"No data available for the region '{region}'")
        return

    # Split data for training and testing
    X_train, X_test, y_train, y_test = train_test_split(region_data[features], region_data[target], test_size=0.2, random_state=42)

    # Dictionary to store results
    model_results = {}

    # Train and compare models
    train_and_compare_models(X_train, X_test, y_train, y_test, model_results)

  

if __name__ == "__main__":
    main()

```
